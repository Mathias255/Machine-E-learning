# 1 Funciones

## 1.1 ¿Qué son?

* Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.
* Las funciones van identificadas por un nombre.
* Se debe usar este nombre para invocar (ejecutar) el código de la función.
* Las funciones calculan un resultado (output) en base a unos parámetros de entrada (input).
* En cada invocación de la función, los parámetros de entrada pueden variar.

## 1.4 Programación de funciones en Python

Formato general:

    statements

Muy comúnmente:

def name(arg1, arg2, ..., argN):
    ...
    return value

## 1.5 Paso de parámetros

* Tipos simples (inmutables) por **valor**: int, float, string.
    - El efecto es como si se creara una copia dentro de la función, aunque realmente no es esto lo que ocurre.
    - Los cambios dentro de la función no afectan fuera.

```python
def cambiar_valor(x):
    x += 3
    print('Dentro de la función:', x)  # x ahora referencia a un nuevo objeto (el 3), pero esto no afecta a la variable de fuera de la función, que sigue apuntando al 2.

    return

a = 2

cambiar_valor(a)
print('Fuera de la función:', a)
```

* Tipos complejos (mutables) por referencia: list, set, dictionary
  - Dentro de la función se maneja el mismo objeto que se ha pasado desde fuera.
  -  Los cambios dentro de la función sí afectan fuera.

```python
def anyadir_2_a(x):
    x.append(4)
            
lista = [0, 1,3]
anyadir_2_a(lista)
print(lista)
```

```python
# with tuples
def anyadir_2_a(x):
    print(id(x))
    return
    
lista = (0, 1)
print(id(lista))
anyadir_2_a(lista)
print(lista)
```

* Si quiero modificar un objeto que es de un tipo básico, los cuales se pasan por valor, puedo
simplemente devolver el resultado de la función y hacer una asignación:

```python
def multiplicar_por_2(x):
    x = x * 2
    return x

a = 3
a = multiplicar_por_2(a)
print(a)
```

* Si quiero que no se modifique un objeto de un tipo complejo, los cuales se pasan por referencia,
puedo pasar una copia a la función:

```python
def anyadir_2_a(x):
    x.append(2)
    print('Dentro de la función: ', x)

lista = [0, 1]
anyadir_2_a(lista.copy())
print('Fuera de la función: ', lista)
```

```python
# Alternativa para obtener una copia: slicing.
lista = [0, 1]
anyadir_2_a(lista[:])
print('Fuera de la función: ', lista)
```

* Si la lista tiene sublistas anidadas hay que hacer un deep copy

```python
import copy
lista = [2, 4, 16, 32, [34, 10, [5,5]]]

# # copia = lista.copy()

copia = copy.deepcopy(lista)
copia[0] = 454
copia[4][2][0] = 64
print(lista)
print(copia)
print(f"{id(lista)} - {id(copia)}")
print(f"{id(lista[4])} - {id(copia[4])}")
```

* Reasignar un parámetro nunca afecta al objeto de fuera:

```python
def funcion_poco_util(x):
    print(id(x))
    x = [2, 3]
    x.append(4)
    print(id(x))

lista = [0, 1]
print(id(lista))
funcion_poco_util(lista)
print(lista)
```

## 1.7 Recursividad

* Funciones que se llaman a sí mismas


```python
def factorial(n):
    if n == 0:
       return 1
    else:
       return n * factorial(n-1)
        
print(factorial(10))

def fact(n):
    a = 1
    for i in range(1, n+1):
        a *= i
    return a
    
print(fact(10))
```

```python
fact = lambda x: 1 if x == 0 else x * fact(x-1)
fact(10)
```

## 1.8 Funciones Lambda

* Es posible definir funciones anónimas (sin nombre).
* Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia *def* no puede (por ejemplo, dentro de una lista o como parámetro de una función).
* Útiles cuando se quiere pasar una función como argumento a otra.

Formato general:

lambda \<lista de argumentos\> : \<valor a retornar\>

* Importante: \<valor a retornar\> no es un conjunto de instrucciones. Es simplemente una expresión *return* que omite esta palabra.
* Las funciones definidas con *def* son más generales:
    - Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con *def*, pero no viceversa.

```python
# Ejemplo: función suma.

def suma(x, y, z):
    return x + y + z
    
print(suma(1, 2, 3))

# # Ejemplo: función suma como expresión lambda.
suma2 = lambda x, y, z : x + y + z
print(suma2(1, 2, 3))
```

```python
# La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizarlas comparaciones.
# Ejemplo: ordenar strings por número de carácteres.

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

cities.sort()
print(cities)

cities.sort(key = lambda x : len(x))
print(cities)

cities.sort(key = len)
print(cities)

def longitud(x):
    return len(x)
    
cities.sort(key = longitud)
print(cities)
```

```python
# Mismo ejemplo sin lambda

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

def num_caracteres(x):
    return len(x)
    
cities.sort(key = num_caracteres)
print(cities)
```

* Nos podemos guardar una referencia para utilización repetida.

```python
cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']

f = lambda x : len(x)

for s in cities:
    print(f(s))
```

* Las expresiones lambda pueden formar parte de una lista.

```python
# Lista de lambdas para mostrar las potencias de 2.
pows = [lambda x: x ** 0,
        lambda x: x ** 1,
        lambda x: x ** 2,
        lambda x: x ** 3,
        lambda x: x ** 4]

for f in pows:
    print(f(2))
```

* Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
* Ideal para escoger entre varias opciones

```python
def switch_dict(operator, x, y):
# # crea un diccionario de opciones y selecciona 'operator'
  return {
      'add': lambda: x + y,
      'sub': lambda: x - y,
      'mul': lambda: x * y,
      'div': lambda: x / y,
}.get(operator, lambda: None)() # retorna None si no encuentra 'operator'

print(switch_dict('add',2,2))
print(switch_dict('div',10,2))
print(switch_dict('mod',17,3))
```

## 1.9 Scopes

* La localización de una asignación a una variable en el código determina desde donde puedes acceder a esa variable.
* Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
* El alcance de la visibilidad de una variable determina su *scope*.
* El término *scope* hace referencia a un espacio de nombres (*namespace*). Por ejemplo, una función establece su propio *namespace*.

```python
X = 10

def func():
    X = 20   # Local a la función. Es una variable diferente, no visible fuera de 'func'

def func_int():
    X = 30
    print(X)
    
func_int()

func()
```

**Regla LEGB** Dada una función *F*, tenemos los siguientes *scopes*:

* **L**ocal: variables locales a *F*.
* **E**nclosing: variables localizadas en funciones que contienen a (o por encima de) *F*.
* **G**lobal: variables definidas en el módulo (fichero) que no están contenidas en ninguna función. Este *scope* no abarca más de un fichero.
* **B**uilt-ins: proporcionadas por el lenguaje.

La resolución de nombres ocurre de abajo a arriba.

```python
# Global (X y func)
X = 99

def func(Y): # Local (Y y Z)
    Z = X + Y
    return Z

print(func(1))
```

```python
X = 1

def func():
   X = 2  # X es una variable diferente

func()
print(X)
```

* La sentencia global nos permite modificar una variable global desde dentro de una función.

```python
X = 1

def func():
    global X
    X = 2

func()
print(X)
```

* No hay necesidad de usar global para referenciar variables. Sólo es necesario para modificarlas.

```python
y, z = 1, 2

def todas_globales():
    global x
    x = y + z # Las tres variables son globales.

todas_globales()
print(x)
print(y)
print(z)
```

```python
X = 77

def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)

    f2()

f1()
```

## Closures

* Las funciones pueden recordar su enclosing scope, independientemente de si éstos continuan
existiendo o no.


```python
def f1():
    X = 88 # Enclosing scope (para f2)
    def f2():
        print(X)
    return f2 # Devuelve f2, sin invocarla
    
accion = f1()
accion()
```

* Sentencia nonlocal para modificar variables que no son locales, pero tampoco globales.

```python
def f1():
    contador = 10
    def f2():
        nonlocal contador
        contador += 1
        print(contador)
    return f2
    
accion = f1()
accion()
accion()
accion()
```

## 1.10.1 Resueltos

```python
# 1 . Ejercicio

# Función que recibe una lista de números
def cuadrados(lista):
    
    # Lista donde se almacenarán los cuadrados
    resultado = []
    
    # Recorrer cada número de la lista de entrada
    for numero in lista:
        
        # Calcular el cuadrado del número y agregarlo a la lista resultado
        resultado.append(numero ** 2)
    
    # Devolver la lista con los cuadrados
    return resultado


# Lista de ejemplo
numeros = [1, 2, 3, 4, 5]

# Llamar a la función y mostrar el resultado
print(cuadrados(numeros))
```

```python
# 2 . Ejercicio

# Función que recibe una cantidad variable de números
def sumar(*numeros):
    
    # Variable para almacenar la suma
    total = 0
    
    # Recorrer todos los números recibidos
    for numero in numeros:
        total += numero
    
    # Devolver la suma total
    return total


# Ejemplos de uso
print(sumar(5, 10))
print(sumar(1, 2, 3, 4, 5))
print(sumar(10, 20, 30, 40))
```

```python
# 3. Ejercicio

# Función que recibe un texto y lo devuelve al revés
def invertir_texto(texto):
    
    # Variable para guardar el texto invertido
    texto_invertido = ""
    
    # Recorrer el texto desde el final hasta el inicio
    for i in range(len(texto) - 1, -1, -1):
        texto_invertido += texto[i]
    
    # Devolver el texto invertido
    return texto_invertido


# Ejemplo de uso
palabra = "hola"
print(invertir_texto(palabra))
```

```python
# 4. Ejercicio 

# Función lambda que invierte un texto
invertir = lambda texto: texto[::-1]

# Ejemplo de uso
palabra = "hola"
print(invertir(palabra))
```

```python
# 5. Ejercicio

# Función que verifica si un número está dentro de un rango
def esta_en_rango(numero, minimo, maximo):
    
    # Comprobar si el número está entre el mínimo y el máximo
    if minimo <= numero <= maximo:
        return True
    else:
        return False


# Ejemplos de uso
print(esta_en_rango(5, 1, 10))   # True
print(esta_en_rango(15, 1, 10))  # False
print(esta_en_rango(7, 5, 8))    # True
```

```python
# 6. Ejercicio 

# Función que devuelve los 5 primeros múltiplos de un número
def primeros_multiplos(numero):
    
    # Verificar que el número sea positivo
    if numero <= 0:
        print("Error: el número debe ser un entero positivo.")
        return []
    
    # Lista para almacenar los múltiplos
    multiplos = []
    
    # Generar los 5 primeros múltiplos
    for i in range(1, 6):
        multiplos.append(numero * i)
    
    # Devolver la lista de múltiplos
    return multiplos


# Ejemplos de uso
print(primeros_multiplos(3))
print(primeros_multiplos(5))
print(primeros_multiplos(0))
```

```python
# 7. Ejercicio

# Función que verifica si una lista tiene elementos duplicados
def tiene_duplicados(lista):
    
    # Recorrer cada elemento de la lista
    for elemento in lista:
        
        # Contar cuántas veces aparece el elemento
        if lista.count(elemento) > 1:
            return True
    
    # Si no se encontraron duplicados
    return False


# Ejemplos de uso
print(tiene_duplicados([1, 2, 3, 4, 5]))
print(tiene_duplicados([1, 2, 3, 2, 5]))
print(tiene_duplicados(["a", "b", "c", "a"]))
```

```python
# 8. Ejercicio

# Función lambda que verifica si una lista tiene duplicados
tiene_duplicados = lambda lista: len(lista) != len(set(lista))

# Ejemplos de uso
print(tiene_duplicados([1, 2, 3, 4, 5]))
print(tiene_duplicados([1, 2, 3, 2, 5]))
print(tiene_duplicados(["a", "b", "c", "a"]))
```

```python
# 9. Ejercicio

# Función que verifica si un texto es un palíndromo
def es_palindromo(texto):
    
    # Convertir el texto a minúsculas
    texto = texto.lower()
    
    # Comparar el texto original con su versión invertida
    if texto == texto[::-1]:
        return True
    else:
        return False


# Ejemplos de uso
print(es_palindromo("reconocer"))
print(es_palindromo("python"))
print(es_palindromo("Radar"))
```
