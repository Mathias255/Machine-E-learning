* ‘linewidth’: brocha
* ‘alpha’: transparencia

Otros: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# Horizontal bar
fig, ax = plt.subplots()
# Example data
people = ('PhD', 'MSc', 'BSc', 'N/A')
y_pos = np.arange(len(people))
percentage = [5,45,25,25]

# print(y_pos)
ax.barh(y_pos, percentage, align='center',
        color='blue', edgecolor='red')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
#ax.invert_yaxis() # labels read top-to-bottom
ax.set_xlabel('Percentage of workforce')
ax.set_title('Level of education in the company')
plt.show()
```

```python
ls_count = (284487, 244560, 208493, 196764)
cities = ('Madrid', 'Barcelona', 'Sevilla', 'Valencia')
data = np.array(ls_count)/1000
y_pos = np.arange(data.size)

width = 0.35 # the width of the bars
opacity = 0.7

fig, ax = plt.subplots()
rects1 = ax.barh(y_pos, data, width, alpha=opacity)

rects1[3].set_color('#872f29')
rects1[2].set_color('#ffcb59')
rects1[1].set_color('#d6ee39')
rects1[0].set_color('#6ade30')

ax.set_yticks(y_pos)
ax.set_yticklabels(cities)

# add some text for labels, title and axes ticks
ax.set_ylabel('Ciudades')
ax.set_xlabel('Km de carril bici')

ax.legend((rects1[0], rects1[1], rects1[2], rects1[3]), cities)
plt.show()
```

## 6 frame.plot()
frame.plot(x,y,label=<Series>,ax,style,kind,xticks,yticks,title,subplots)
* (solo para Series) label: referencia para la leyenda
* ax: subfigura en la que dibujar los datos
* style: estilo de la línea
* kind: tipo de gráfico (’bar’, ‘pie’, ‘hist’, ‘area’ ‘line’, ‘barh’, ‘density’, ‘kde’).
* xticks y yticks: valores en los ejes X e Y
* title: string como título
* (solo para DataFrame) subplots: bool si se desean subfiguras separadas para cada columna.
* (solo para DataFrame) ‘x’ e ‘y’ se pueden utilizar para seleccionar columnas para cada eje

```python
#scatter de clases, coloreando segun clase en matplotlib
iris = sns.load_dataset('iris')

# preparacion de los datos
data = []
colors = ["red", "green", "blue"]
groups = ["setosa", "versicolor", "virginica"]
columns = ["petal_length", "petal_width"]
for i in groups:
    # busqueda por clase definida
    data.append(iris[iris["species"] == i][columns].values)


#creacion de la grafica
fig = plt.figure()
ax = fig.add_subplot()
for count, d in enumerate(data):
    # separar las variables del grafico
    x, y = d[:,0], d[:,1]
    ax.scatter(x, y, alpha = 0.4, c = colors[count -1], edgecolors = 'none', s = 30, label = groups[count - 1])
#etiquetas
ax.set_xlabel(columns[0])
ax.set_ylabel(columns[1])
plt.legend(loc='best')
plt.title('Matplot scatter plot')
plt.show()
```

## 8.1 matplotlib
* crear una lista de años de 2000 a 2020 (eje x)
* generar datos aleatorios para el eje y (hacerlo 4 veces distintas: y0, y1, y2 e y3)
* Representar las 4 secuencias de datos aleatorios en una sola figura utilizando matplotlib
* Añade una leyenda para poder identificar cada secuencia en el gráfico
* Nombra las secuencias de la siguiente forma: “Hawaii”, “San Marino”, “Islas Feroe”,
“Guayana”
* Representar los mismos datos en 4 subfiguras (axes) distintas, parte de la misma figura
* Añade la siguiente info:
  
       – Título: “Datos aleatorios”
       – Leyenda
       – Nombre del eje x: “Año”
       – Nombre del eje y: “GDP”
  
* Añade al menos una anotación (texto y flecha) a uno de los gráficos


```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Crear la lista de años de 2000 a 2020 para el eje X (21 elementos)
eje_x = list(range(2000, 2021))

# 2. Generar datos aleatorios para el eje Y (4 veces distintas)
# Usamos una semilla para que los datos sean reproducibles, puedes quitarla si quieres
np.random.seed(42) 
y0 = np.random.randint(10, 100, size=21)
y1 = np.random.randint(10, 100, size=21)
y2 = np.random.randint(10, 100, size=21)
y3 = np.random.randint(10, 100, size=21)

# Guardamos los nombres asignados
nombres = ["Hawaii", "San Marino", "Islas Feroe", "Guayana"]

# =====================================================================
# FIGURA 1: Las 4 secuencias en una sola figura
# =====================================================================
plt.figure(figsize=(10, 6))

plt.plot(eje_x, y0, marker='o', label=nombres[0])
plt.plot(eje_x, y1, marker='s', label=nombres[1])
plt.plot(eje_x, y2, marker='^', label=nombres[2])
plt.plot(eje_x, y3, marker='x', label=nombres[3])

# Añadir información requerida
plt.title("Datos aleatorios - Todo en una figura", fontsize=14, fontweight='bold')
plt.xlabel("Año", fontsize=12)
plt.ylabel("GDP", fontsize=12)
plt.xticks(range(2000, 2021, 2))
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)

# =====================================================================
# FIGURA 2: Las 4 secuencias en 4 subfiguras (axes) distintas
# =====================================================================
# Creamos una cuadrícula de 2x2 subgráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 8), sharex=True)
fig.suptitle("Datos aleatorios", fontsize=16, fontweight='bold')

# Agrupamos los datos en listas para iterar cómodamente
datos_y = [y0, y1, y2, y3]
colores = ['blue', 'orange', 'green', 'red']
marcadores = ['o', 's', '^', 'x']

# Aplanar la matriz de subplots (de 2x2 a una lista de 4 elementos) para el bucle
axs_flat = axs.flatten()

for i in range(4):
    ax = axs_flat[i]
    ax.plot(eje_x, datos_y[i], marker=marcadores[i], color=colores[i], label=nombres[i])
    
    # Configuración individual de cada subgráfico
    ax.set_xlabel("Año")
    ax.set_ylabel("GDP")
    ax.legend(loc='upper left')
    ax.grid(True, linestyle='--', alpha=0.5)

# --- Añadir una anotación (texto y flecha) al primer subgráfico (Hawaii) ---
# Buscamos el valor máximo de y0 para apuntar a él de forma dinámica
idx_max = np.argmax(y0)
anio_max = eje_x[idx_max]
valor_max = y0[idx_max]

axs_flat[0].annotate(
    'Pico más alto', 
    xy=(anio_max, valor_max),          # Punto al que apunta la flecha
    xytext=(anio_max - 4, valor_max - 15), # Ubicación del texto
    arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6) # Estilo de la flecha
)

# Ajustar diseño para que no se encabalguen los títulos ni etiquetas
plt.tight_layout()

# Mostrar ambos gráficos
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Cargar el dataset desde la URL en formato 'Raw'
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
df = pd.read_csv(url)

# Limpiamos valores nulos para evitar errores en las gráficas numéricas
df = df.dropna()

# =====================================================================
# PUNTO 1: Relación de cada pareja de atributos (Pairplot)
# =====================================================================
# Representa visualmente la relación de todas las combinaciones de variables numéricas
sns.pairplot(df, hue="species", corner=True)
plt.suptitle("Relación de cada pareja de atributos", y=1.02, fontsize=14)
plt.show()

# =====================================================================
# PUNTO 2: Correlación entre dos variables continuas y numéricas
# =====================================================================
plt.figure(figsize=(8, 5))
# Escogemos la longitud del ala (flipper_length_mm) y la masa corporal (body_mass_g)
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g")
plt.title("Correlación: Longitud de Aleta vs Masa Corporal (Correlación Positiva)")
plt.xlabel("Longitud de la Aleta (mm)")
plt.ylabel("Masa Corporal (g)")
plt.grid(True, alpha=0.3)
plt.show()

# =====================================================================
# PUNTO 3: Dos variables numéricas en base a otra categórica
# =====================================================================
plt.figure(figsize=(8, 5))
# x e y numéricas (bill_length_mm, bill_depth_mm) separadas por color (hue) usando la especie
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species", palette="Set1")
plt.title("Dimensiones del Pico segmentadas por Especie")
plt.xlabel("Longitud del Pico (mm)")
plt.ylabel("Profundidad del Pico (mm)")
plt.show()

# =====================================================================
# PUNTO 4.1: Distribución de valores (media, distribución, outliers)
# =====================================================================
# Un Boxplot es el gráfico ideal para ver la media (mediana), cuartiles y outliers
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="body_mass_g", color="lightblue")
plt.title("Distribución, Mediana y Outliers de Masa Corporal")
plt.xlabel("Masa Corporal (g)")
plt.show()

# =====================================================================
# PUNTO 4.2: Comparar distribución de dos variables (Densidad / KDE)
# =====================================================================
plt.figure(figsize=(8, 5))
# Comparamos cuantitativamente la densidad de la longitud y profundidad del pico
sns.kdeplot(data=df, x="bill_length_mm", fill=True, label="Longitud del Pico", alpha=0.5)
sns.kdeplot(data=df, x="bill_depth_mm", fill=True, label="Profundidad del Pico", alpha=0.5)
plt.title("Comparación de Densidades (KDE)")
plt.xlabel("Medida (mm)")
plt.ylabel("Densidad")
plt.legend()
plt.show()

# =====================================================================
# PUNTO 4.3: Regresión Lineal de Orden > 1 (Polinomial) vs Orden == 1
# =====================================================================
# Para demostrar esto de forma clara, la relación entre la longitud del pico
# y su profundidad a nivel global (sin separar por especie) no es del todo lineal.
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico Izquierdo: Orden == 1 (Regresión lineal simple)
sns.regplot(ax=axes[0], data=df, x="bill_length_mm", y="bill_depth_mm", 
            order=1, line_kws={"color": "red"}, scatter_kws={"alpha": 0.5})
axes[0].set_title("Regresión Lineal Simple (Orden = 1)")
axes[0].set_xlabel("Longitud del Pico (mm)")
axes[0].set_ylabel("Profundidad del Pico (mm)")

# Gráfico Derecho: Orden > 1 (Regresión Polinomial de Orden 2)
# Al tener un comportamiento con curvaturas debido al cruce de especies, el orden 2 se adapta mejor a la masa global de datos
sns.regplot(ax=axes[1], data=df, x="bill_length_mm", y="bill_depth_mm", 
            order=2, line_kws={"color": "green"}, scatter_kws={"alpha": 0.5})
axes[1].set_title("Regresión Polinomial (Orden = 2)")
axes[1].set_xlabel("Longitud del Pico (mm)")
axes[1].set_ylabel("Profundidad del Pico (mm)")

plt.tight_layout()
plt.show()
```
