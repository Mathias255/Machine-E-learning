```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

# Fijar semilla para reproducibilidad
random.seed(42)

# Crear lista con los niveles de estudio y sus frecuencias
nivel_estudios = []
nivel_estudios.extend(['Secundaria'] * 110)
nivel_estudios.extend(['Bachillerato'] * 172)
nivel_estudios.extend(['Grado'] * 145)
nivel_estudios.extend(['Master'] * 52)
nivel_estudios.extend(['Doctorado'] * 21)

# Mezclar los datos de forma aleatoria
nivel_estudios = np.random.permutation(nivel_estudios)

# Crear un DataFrame de Pandas
df = pd.DataFrame({'Nivel_Estudios': nivel_estudios})

# Mostrar las primeras filas del DataFrame
print("Primeros registros:")
print(df.head())

# Calcular la frecuencia absoluta de cada nivel de estudios
frecuencias = df['Nivel_Estudios'].value_counts()

# Mostrar la tabla de frecuencias
print("\nFrecuencia absoluta:")
print(frecuencias)

# Obtener las categorías (niveles de estudio)
estudios = frecuencias.index

# Obtener los valores de frecuencia
frecuencia = frecuencias.values

# ==========================
# DIAGRAMA DE BARRAS
# ==========================

# Definir el tamaño de la figura
plt.figure(figsize=(8, 5))

# Crear el gráfico de barras
plt.bar(estudios, frecuencia)

# Agregar etiquetas y título
plt.xlabel('Nivel de estudios')
plt.ylabel('Frecuencia absoluta')
plt.title('Diagrama de barras')

# Mostrar el gráfico
plt.show()

# ==========================
# DIAGRAMA DE SECTORES
# ==========================

# Definir colores para cada sector
colores = ['red', 'blue', 'orange', 'green', 'cyan']

# Definir el tamaño de la figura
plt.figure(figsize=(8, 8))

# Crear el gráfico circular
plt.pie(
    frecuencia,              # Datos
    labels=estudios,         # Etiquetas
    colors=colores,          # Colores
    autopct='%1.2f%%'        # Mostrar porcentaje con 2 decimales
)

plt.title('Diagrama de sectores')

plt.show()
```
